```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('output',axis=1)
y=df['output']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=500)
```

```python
lr.fit(X_train,y_train)
```

```python
pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
classification_report(y_test,pred)
```

```python
confusion_matrix(y_test,pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
df.describe().loc['std'].mean()
```

```python
kn = KNeighborsClassifier(n_neighbors=7)
```

```python
kn.fit(X_train,y_train)
```

```python
knpred = kn.predict(X_test)
```

```python
classification_report(y_test,knpred)
```

```python
confusion_matrix(y_test,knpred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rnd = RandomForestClassifier()
```

```python
rnd.fit(X_train,y_train)
```

```python
rnpred = rnd.predict(X_test)
```

```python
classification_report(y_test,rnpred)
```

```python
confusion_matrix(y_test,rnpred)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
sv = SVC()
```

```python
sv.fit(X_train,y_train)
```

```python
svpred = sv.predict(X_test)
```

```python
classification_report(y_test,svpred)
```

```python
confusion_matrix(y_test,svpred)
```

```python
q = { 'C':[0.1,1,10,100,1000], 'gamma':[1,0.1,0.01,0.001,0.0001,0.00001]}
grid = GridSearchCV(SVC(),q,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grpred = grid.predict(X_test)
```

```python
classification_report(y_test,grpred)
```

```python
confusion_matrix(y_test,grpred)
```

```python
#so Random Forest Gave the best precision
```
